## WEB SCRAPING

Web scraping is the automated extraction of data from websites. It involves parsing HTML pages, extracting specific information, and storing it for further analysis. It's important as it enables access to vast amounts of data online, allowing businesses, researchers, or developers to gather valuable information from various websites efficiently.

-- The given Python code scrapes information for a book list from the Amazon website using BeautifulSoup and Selenium. To start, the script imports the necessary libraries, such as BeautifulSoup for HTML parsing and Selenium for web automation. It defines the URL of Amazon as well as the function scrape_book_details, which is in charge of going to Amazon, searching for a certain book, selecting the first result, and collecting information from it such the book's title, rating, and user count.

-- The script then iterates over a list of book names, using the scrape_book_details function for each book as it goes. A list called all_book_details contains the findings, including book details. The script then prints out each book's data, including the title, rating, and total number of user ratings, in an easily accessible style.

-- It's important to remember that the script depends on the Chromedriver executable since it uses the Chrome browser for web automation. Before executing the script, we also need to confirm that the required Python libraries (chromedriver_autoinstaller, beautifulsoup4, and selenium) are installed. Furthermore, we must confirm that the Chromedriver version we have installed is compatible with our Chrome browser. The script is a useful illustration of web scraping book data from an online retailer such as Amazon.

```python
# Importing necessary libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

"""
Scrape details for a specific book.

"""

# Defining the URL for Amazon
url = "https://www.amazon.com"

# Function to scrape book details
def scrape_book_details(url,book_name):
    # Creating a new instance of the Chrome driver
    driver = webdriver.Chrome()
    url = url

    try:
        driver.get(url)

        # Finding the search box using its name attribute
        search_box = driver.find_element(By.NAME, "field-keywords")
        search_box.clear()
        search_box.send_keys(book_name)

        # Submitting the search form
        search_box.submit()

        # Waiting for a while to let the results load (you might want to use WebDriverWait for more robust waiting)
        driver.implicitly_wait(10)

        # Clicking on the first search result
        first_result = driver.find_element(By.XPATH, '//a[@class="a-link-normal s-no-outline"]')
        first_result.click()

        driver.implicitly_wait(15)

        # Getting the HTML content
        html = driver.page_source

        # Parsing HTML content with BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")

        # Extracting book details
        title = soup.find("span", id="productTitle").text.strip()
        rating = soup.find("span", class_="a-icon-alt").text.strip()
        num_user_rated = soup.find("span", id="acrCustomerReviewText").text.strip()

        # Returning the extracted details
        return {"title": title, "rating": rating, "num_user_rated": num_user_rated}

    finally:
        # Closing the browser window
        driver.quit()



# Listing of books to scrape
book_list = [
    "The Big Book of American Trivia",
    "Lay Down my Sword and Shield by James Lee Burke",
    "Hamlet (Wordsworth Classics)",
    "Calculating God: A Novel",
    "Plan of Attack",
    "Arcadia book by Tom Stoppard",
    "Le Colonel Chabert by Honoré de Balzac",
    "The Lake of Darkness (Vintage Crime/Black Lizard) by Ruth Rendell",
    "The Run by Stuart Woods",
    "Pandora by Anne Rice",
    "Self Analysis by L. Ron Hubbard"
]

# Initializing an empty list to store book details
url = "https://www.amazon.com"
all_book_details = []
for book_name in book_list:
    book_details = scrape_book_details(url,book_name)
    all_book_details.append(book_details)

# Displaying the details for each book
for i, book_details in enumerate(all_book_details, start=1):
    print(f"\nDetails for Book {i} - {book_list[i-1]}:")
    print(f"Title: {book_details['title']}")
    print(f"Rating: {book_details['rating']}")
    print(f"num_user_rated: {book_details['num_user_rated']}")
```

```python
# Importing the pandas library
import pandas as pd

# Step 1: Calculating Average Ratings
# Grouping the DataFrame 'df' by 'Book-Author' and calculating the mean of 'Book-Ratings' for each author
average_ratings = df.groupby('Book-Author')['Book-Ratings'].mean()

# Step 2: Count Number of Books per Author
# Counting the number of books for each author in the 'Book-Author' column
book_counts = df['Book-Author'].value_counts()

# Combining the data into a new DataFrame
# Creating a new DataFrame 'author_data' with columns for 'Average Rating' and 'Book Count'
author_data = pd.DataFrame({'Average Rating': average_ratings, 'Book Count': book_counts})

author_data.head()
```

## Surprise Feature: (New Feature which we included)

```python
# Importing necessary libraries
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import SVD

# Creating a label encoder
label_encoder = LabelEncoder()

# Filtering out rows with 'Book-Ratings' equal to 0
obj = df[df['Book-Ratings'] != 0]

# Encode book titles and authors
obj['Year-Of-Publication'] = label_encoder.fit_transform(obj['ISBN'])
obj['Book-Author'] = label_encoder.fit_transform(obj['Book-Author'])

# Visualizing the distribution of ratings
plt.hist(obj['Book-Ratings'], bins=10, edgecolor='black')
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Scale ratings to a range between 1 and 10 (or any desired range)
scaler = MinMaxScaler(feature_range=(1, 10))
obj['Book-Ratings'] = scaler.fit_transform(obj['Book-Ratings'].values.reshape(-1, 1))

# Loading dataset into Surprise
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(obj[['Year-Of-Publication', 'Book-Ratings', 'Book-Author']], reader)

# Splitting the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2)

# Choosing the SVD algorithm
algo = SVD()

# Training the algorithm on the trainset
algo.fit(trainset)
```
